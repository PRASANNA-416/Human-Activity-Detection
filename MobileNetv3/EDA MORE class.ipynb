```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from PIL import Image

# Path of the main directory
main_directory = "C:/Users/Prasanna P M/Human Project/Project U Turn/MORETRAIN/"

# Data augmentation parameters
augmentation_params = {
    'rotation_range': 20,
    'width_shift_range': 0.2,
    'height_shift_range': 0.2,
    'shear_range': 0.2,
    'zoom_range': 0.2,
    'horizontal_flip': True
}

# Target image size
target_size = (224, 224)

# Iterate through each subdirectory
for subdirectory in os.listdir(main_directory):
    subdirectory_path = os.path.join(main_directory, subdirectory)
    
    if os.path.isdir(subdirectory_path):
        # Create an ImageDataGenerator with augmentation parameters
        datagen = ImageDataGenerator(**augmentation_params)
        
        # Get a list of image files within the subdirectory
        image_files = [filename for filename in os.listdir(subdirectory_path) if os.path.isfile(os.path.join(subdirectory_path, filename))]
        
        # Iterate through each image file and apply data augmentation
        for image_file in image_files:
            image_path = os.path.join(subdirectory_path, image_file)
            
            # Load the image
            image = Image.open(image_path)
            
            # Resize the image to the target size
            image = image.resize(target_size)
            
            # Convert the image to a NumPy array
            image_array = np.array(image)
            
            # Generate augmented image
            augmented_image = datagen.flow(image_array.reshape(1, *image_array.shape), batch_size=1).next()[0].astype(np.uint8)
            
            # Convert augmented image back to PIL image
            augmented_image = Image.fromarray(augmented_image)
            
            # Save the augmented image
            augmented_image_path = os.path.join(subdirectory_path, "augmented_" + image_file)
            augmented_image.save(augmented_image_path)

print("Data augmentation applied successfully!")
```

```python
import os
import numpy as np
import tensorflow as tf
from PIL import Image

# Path of the main directory
main_directory = "C:/Users/Prasanna P M/Human Project/Project U Turn/MORETRAIN/"

# Data augmentation parameters
augmentation_params = {
    'brightness': 0.4,
    'contrast': 1.0,
    'jpeg_quality': [80, 100],
    'saturation': 0.5
}

# Target image size
target_size = (224, 224)

# Iterate through each subdirectory
for subdirectory in os.listdir(main_directory):
    subdirectory_path = os.path.join(main_directory, subdirectory)
    
    if os.path.isdir(subdirectory_path):
        # Get a list of image files within the subdirectory
        image_files = [filename for filename in os.listdir(subdirectory_path) if os.path.isfile(os.path.join(subdirectory_path, filename))]
        
        # Iterate through each image file and apply data augmentation
        for image_file in image_files:
            image_path = os.path.join(subdirectory_path, image_file)
            
            # Load and preprocess the image
            image = Image.open(image_path)
            image_tensor = tf.convert_to_tensor(np.array(image))
            image_tensor = tf.image.resize(image_tensor, target_size)
            image_tensor = tf.cast(image_tensor, tf.float32) / 255.0  # Normalize pixel values
            
            # Apply data augmentation
            augmented_image = tf.image.stateless_random_brightness(image_tensor, max_delta=augmentation_params['brightness'], seed=(1, 2))
            augmented_image = tf.image.stateless_random_contrast(augmented_image, lower=1 - augmentation_params['contrast'], upper=1 + augmentation_params['contrast'], seed=(1, 2))
            augmented_image = tf.image.stateless_random_jpeg_quality(augmented_image, min_jpeg_quality=augmentation_params['jpeg_quality'][0], max_jpeg_quality=augmentation_params['jpeg_quality'][1], seed=(1, 2))
            augmented_image = tf.image.stateless_random_saturation(augmented_image, lower=1 - augmentation_params['saturation'], upper=1 + augmentation_params['saturation'], seed=(1, 2))
            
            # Convert augmented image back to PIL image
            augmented_image = tf.clip_by_value(augmented_image * 255.0, 0.0, 255.0)
            augmented_image = tf.cast(augmented_image, tf.uint8)
            augmented_image = Image.fromarray(augmented_image.numpy())
            
            # Generate a new file name for the augmented image
            augmented_image_file = "col_" + image_file
            
            # Save the augmented image with the new file name
            augmented_image_path = os.path.join(subdirectory_path, augmented_image_file)
            augmented_image.save(augmented_image_path)
```

```python
import os
import random
import shutil
from math import ceil

def move_images(original_directory, duplicate_directory):
    # Iterate through each subdirectory in the original directory
    for subdir in os.listdir(original_directory):
        original_subdir_path = os.path.join(original_directory, subdir)
        duplicate_subdir_path = os.path.join(duplicate_directory, subdir)
        
        # Create the corresponding subdirectory in the duplicate directory
        os.makedirs(duplicate_subdir_path, exist_ok=True)
        
        # Get the list of image files in the original subdirectory
        image_files = [file for file in os.listdir(original_subdir_path) if file.endswith('.jpg') or file.endswith('.png')]
        
        # Calculate the number of images to move (10% of total images rounded up)
        num_images_to_move = ceil(len(image_files) * 0.1)
        
        # Choose a random selection of images to move
        images_to_move = random.sample(image_files, num_images_to_move)
        
        # Move the selected images to the duplicate subdirectory
        for file in images_to_move:
            original_file_path = os.path.join(original_subdir_path, file)
            duplicate_file_path = os.path.join(duplicate_subdir_path, file)
            shutil.move(original_file_path, duplicate_file_path)

    print("Image movement completed successfully.")

# Provide the paths for the original directory and duplicate directory
original_directory = "C:/Users/Prasanna P M/Human Project/Project U Turn/MORETRAIN/"
duplicate_directory = "C:/Users/Prasanna P M/Human Project/Project U Turn/MOREVALID/"

# Call the function to move the images
move_images(original_directory, duplicate_directory)
```

```python
import os
import cv2
import shutil

def convert_to_grayscale(image_path):
    # Read the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Expand the dimensions to match the shape (224, 224, 1)
    gray_image = gray_image[:, :, None]

    return gray_image

def stack_channels(image_path, gray_image):
    # Read the original image
    original_image = cv2.imread(image_path)

    # Stack the grayscale image channels to create a final image
    final_image = cv2.merge([gray_image, gray_image, gray_image])

    return final_image

def copy_and_convert_images(main_directory, duplicate_directory):
    # Iterate through each subclass directory in the main directory
    for subclass_dir in os.listdir(main_directory):
        subclass_dir_path = os.path.join(main_directory, subclass_dir)
        duplicate_subclass_dir_path = os.path.join(duplicate_directory, subclass_dir)
        
        # Create the corresponding subclass directory in the duplicate directory
        os.makedirs(duplicate_subclass_dir_path, exist_ok=True)
        
        # Get the list of image files in the subclass directory
        image_files = [file for file in os.listdir(subclass_dir_path) if file.endswith('.jpg') or file.endswith('.png')]
        
        # Iterate through each image file
        for file in image_files:
            original_file_path = os.path.join(subclass_dir_path, file)
            duplicate_file_path = os.path.join(duplicate_subclass_dir_path, file)
            
            # Convert the image to grayscale
            gray_image = convert_to_grayscale(original_file_path)
            
            # Stack the grayscale channels to create the final image
            final_image = stack_channels(original_file_path, gray_image)
            
            # Save the final image to the duplicate subclass directory
            cv2.imwrite(duplicate_file_path, final_image)

    print("Image conversion and copying completed successfully.")

# Specify the paths for the main directory and duplicate directory
main_directory = "C:/Users/Prasanna P M/Human Project/Project U Turn/MORETEST/"
duplicate_directory = "C:/Users/Prasanna P M/Human Project/Project U Turn/MORETEST_GRAY_SCALE/"

# Call the function to copy and convert the images
copy_and_convert_images(main_directory, duplicate_directory)
```
